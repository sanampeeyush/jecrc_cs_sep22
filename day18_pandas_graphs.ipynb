#### Graph Functions
1. plot                     - Simple Line Plot
2. plot.area                - Filled Version of Line Plot
3. plot.bar  /  plot.barh   - Bar Graph
4. plot.box                 - Box Plot
5. plot.density             - Density Line (Smooth Curved)
6. plot.hexbin              - Hexagon Version of Scatter with Desnity
7. plot.hist                - Histogram
8. plot.kde                 - Kernel Density Estimator (Density Graph)
9. plot.pie                 - Pie Chart of Categorical Data
10. plot.scatter            - Scatter Graph

<img src="https://i.ytimg.com/vi/BE8CVGJuftI/maxresdefault.jpg">

```python
df2.plot(); plt.title('Line Plot'); plt.show()
df2.plot.area(); plt.title('Area Plot'); plt.show()
df2.plot.bar(); plt.title('Bar Plot'); plt.show()
df2.plot.box(); plt.title('Box Plot'); plt.show()
df2.plot.barh(); plt.title('Bar Horizontal Plot'); plt.show()
df2.plot.density(); plt.title('Desnsity Plot'); plt.show()
df2.plot.kde(); plt.title('KDE Plot'); plt.show()
df2['a'].plot.hist(); plt.title('Histogram Plot'); plt.show()

df2.plot.scatter(x='a', y='b'); plt.title('Scatter Plot'); plt.show()

df2.plot.hexbin(x='a', y='b', C='c', gridsize=6); 
plt.title('Hexbin Plot'); plt.show()
```

# Seaborn

```python
import seaborn as sns
```

##### Categories
1. Categorical Plots (hue can be used | comparison with categorical)
    a. countplot (single param)
    b. barplot (one categorical, one numerical | calculates mean by default)
    c. boxplot
    d. violinplot
    e. swarmplot
2. Distribution Plots
    a. distplot
    b. displot
    c. jointplot
    d. pairplot
    e. kdeplot
    f. rugplot
    g. scatterplot
3. Matrix Plots
    a. heatmap
    b. clustermap
    c. pairplot (same as above one)
4. Regression Plot
    a. lmplot
    b. regplot
5. Grid Plots
    a. PairGrid
    b. JointGrid
    c. FacetGrid

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.info()
```

```python
tips.describe(include='all')
```

```python
tips['day'].value_counts()
```

```python
tips.groupby(['sex','time'])['day'].value_counts().sort_index()
```

```python
sns.countplot(x='day', data=tips)
```

```python
sns.countplot(x='day', data=tips, hue='sex')
```

```python
sns.barplot(x='day', y='total_bill', data = tips)
```

```python
sns.set_style('whitegrid')
```

```python
sns.barplot(x='day', y='total_bill', data = tips, estimator=sum)
```

```python
sns.barplot(x='day', y='total_bill', data = tips, hue='time',
             estimator=np.mean)
```

```python
sns.swarmplot(x='day', y='total_bill', data = tips, hue='smoker',)
plt.show()
sns.boxplot(x='day', y='total_bill', data = tips, hue='smoker',)
plt.show()
sns.violinplot(x='day', y='total_bill', data = tips, hue='smoker',)
plt.show()
```
